```python
import joblib
import cv2
import numpy as np
import time
import pyscreenshot as ImageGrab
```

```python
model = joblib.load('model/digit_recognizer') # Load the saved model
%mkdir img
image_folder = "img/"
```

```python
while True:
    print("Capturing...")
    img = ImageGrab.grab(bbox=(3,185,415,685)) # x, y, w, h

    img.save(image_folder + "img.png") # Save the image as img.png
    im = cv2.imread(image_folder + "img.png") # Read the image
    im_grey = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY) # Convert to greyscale it will reduce the time of processing
    im_grey = cv2.GaussianBlur(im_grey, (15, 15), 0) # Reduce noise and increase smoothness
    
    #Threshold the image
    ret, im_th = cv2.threshold(im_grey.copy(), 90, 255, cv2.THRESH_BINARY_INV) #Focus on the digits and remove the background #This is an extra step to make the image more clear
    roi = cv2.resize(im_th, (28, 28), interpolation=cv2.INTER_AREA) # Resize the image to 28x28 pixels

    rows, cols = roi.shape # Get the height and width of the image

    X = []

    # Add pixel one by one into data array
    for i in range(rows):
        for j in range(cols):
            k = roi[i, j]
            if k > 100:
                k = 1
            else:
                k = 0
            X.append(k)
    # print(X)

    predictions = model.predict([X]) # Make the predictions
    print("Prediction: ", predictions[0]) # Print the predictions
    cv2.putText(im, "Prediction is: " + str(predictions[0]), (20, 20), 0, 0.8, (0, 255, 0), 2, cv2.LINE_AA) # Put the prediction text on the image

    cv2.startWindowThread() # Start a thread to show the image
    cv2.namedWindow("Result") # Create a window to show the image
    cv2.imshow("Result", im) # Show the image
    cv2.waitKey(10000) # Wait for 10 seconds

    if cv2.waitKey(1) == 27 or cv2.waitKey(1) == 13: #27 is the ascii code for ESC key, 13 is the ascii code for Enter key
        break

cv2.destroyAllWindows() # Destroy all the windows
```
